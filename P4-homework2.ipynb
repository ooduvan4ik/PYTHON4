```python
import requests
from bs4 import BeautifulSoup 
from pprint import pprint
import json

url = "https://hh.ru/search/vacancy" #я видела пример, где поиск по api.hh.ru/vanacies - в чем разница? правильно я понимаю, что там другой подход к парсингу?
salary_min=300000
salary_max=400000
city='Москва'
key_words='PYTHON'
page=0
pages=3
per_page=100 
area=1

my_header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 YaBrowser/21.9.1.684 Yowser/2.5 Safari/537.36'}#{'HH-User-Agent':  'api-test-agent' } #HH-User-Agent 'MyApp/1.0 (potapych190@mail.ru)'
my_params = {'city': city, 
             'text': key_words, 
             'salary_from': salary_min, 
             'salary_to': salary_max, 
             'area':area,
             'page':page}

vacancy_list = []

while my_params['page']< pages:

    response=requests.get(url, headers = my_header, params=my_params)
    #page=N&per_page=M

    dom = BeautifulSoup(response.text, 'html.parser')
    vacancies=dom.find_all("div", {'class':'vacancy-serp-item'})

    #print(vacancies)

    if response.ok and vacancies:
        for vac in vacancies:
            vac_data = {}
            try:
                info = vac.find('a', {'class': 'bloko-link'})
                #print(info)
                name = info.text
                #print(name)
                link = info['href']
                #print(link)
            except:
                info = None
            try:
                zp_info = vac.find('div', {'class': "vacancy-serp-item__sidebar"}).text
                zp_info.strip("‪\u202f") # не помогло, почему-то выводится данное сочетание...
                #print(zp_info)
            except:
                zp_info = None    
            
            zp_lower=None #пока не придумала, как обработать красиво, если придумаю, перегружу.
            zp_highest=None
            #print(zp_info)
            web='hh.ru'
              
            vac_data['name'] = name
            vac_data['zp']=zp_info
            vac_data['link'] = link
            vac_data['web'] = web
            
            vacancy_list.append(vac_data)
            #print(vacancy_list)
            my_params['page']+=1 # почему то тут 20 страниц, а не 3 в итоге цикл проходит, почему??
       
            
            
    else:
        break        
            
print(vacancy_list)            
            
            
            
 

```
